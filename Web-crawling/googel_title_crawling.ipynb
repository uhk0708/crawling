```python
from selenium import webdriver as wb
from bs4 import BeautifulSoup as bs
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd
```

```python
driver = wb.Chrome()
driver.set_window_size(800, 600)

Url = 'https://www.google.com/search?q=%ED%8C%8C%EB%AC%98+%ED%8C%A8%EB%9F%AC%EB%94%94&sca_esv=173e16c3c997395a&tbm=vid&prmd=ivnsmbz&sxsrf=ADLYWIL2wp9zB0B1uYcxHLq_Ssj388AGTg:1716710616518&ei=2OxSZoaHH_vm2roPv-6t6AM&start=0&sa=N&ved=2ahUKEwiG0-qz7aqGAxV7s1YBHT93Cz04ChDy0wN6BAgSEAQ&biw=3035&bih=1459&dpr=1.35'
driver.get(Url)
time.sleep(6)



soup = bs(driver.page_source, 'lxml') 

title_list = soup.select(".LC20lb.MBeuO.DKV0Md")
date_list = soup.select("div.gqF9jc > span:nth-child(3)")
site_list = soup.select(".cuFRh")
channel_list = soup.select("div.gqF9jc > span:nth-child(2)") 


data = []

page = 1

while True:  
    soup = bs(driver.page_source, 'lxml') 
    title_list = soup.select(".LC20lb.MBeuO.DKV0Md")
    date_list = soup.select("div.gqF9jc > span:nth-child(3)")
    site_list = soup.select(".cuFRh")
    channel_list = soup.select("div.gqF9jc > span:nth-child(2)")      

    for title, date, site, channel in zip(title_list, date_list, site_list, channel_list): 
             title_text = title.text
             date_text = date.text
             site_text = site.text
             channel_text = channel.text
             if ('파묘' in title_text) & ('패러디' in title_text):
                 print(f'제목 : {title_text}\n날짜: {date_text}\n사이트: {site_text}\n채널명: {channel_text}')
                 print()
                 data.append([title_text, date_text, site_text, channel_text])

    try:
        next_page = driver.find_element(By.CSS_SELECTOR, '#pnnext')
        next_page.click()
        time.sleep(2)
    except Exception as e:
        print("더 이상 페이지가 없습니다. 스크래핑을 종료합니다.")
        break
        


df = pd.DataFrame(data, columns=['제목','날짜', '플랫폼', '채널명'])
df.to_csv('phamo_parody_crawling.csv', index=False, encoding='utf-8-sig')
```
